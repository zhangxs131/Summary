```python
from datasets import load_dataset

dataset=load_dataset('cnn_dailymail',"3.0.0")
dataset
```

```python
dataset['train'][0]
```

```python
import pandas as pd

df=pd.DataFrame(dataset['train'][:10000])
df=df.drop(columns='id')
df.columns = ['text','summary']
df
```

```python
df.to_csv('train.csv')
```

```python
df=pd.DataFrame(dataset['validation'][:1000])
df=df.drop(columns='id')
df.columns = ['text','summary']
df
```

```python
df.to_csv('valid.csv')
```

```python
!python transformers/examples/pytorch/summarization/run_summarization.py \
    --model_name_or_path t5-small \
    --do_train \
    --do_eval \
    --train_file train.csv \
    --validation_file valid.csv \
    --text_column text \
    --summary_column summary \
    --num_train_epochs 1 \
    --source_prefix "summarize: " \
    --output_dir /content/tst-summarization \
    --save_strategy epoch \
    --per_device_train_batch_size=4 \
    --per_device_eval_batch_size=4 \
    --overwrite_output_dir \
    --predict_with_generate
```

```python
from transformers import T5ForConditionalGeneration,AutoTokenizer
from  transformers import pipeline

model=T5ForConditionalGeneration.from_pretrained('/content/tst-summarization')
tokenizer=AutoTokenizer.from_pretrained('/content/tst-summarization')

generator=pipeline('summarization',model=model,tokenizer=tokenizer)

text=dataset['test'][6]['article']

result=generator(text,min_length=10,max_length=40)
print(text)
result
```
